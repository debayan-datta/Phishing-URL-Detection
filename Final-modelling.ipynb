```python
url.replace('?', -1, inplace=True)
url.replace('>3', -1, inplace=True)
```

```python
non_numeric_columns = []
for column in url.columns:
    if pd.to_numeric(url[column], errors='coerce').isnull().any():
        non_numeric_columns.append(column)

print("Columns with non-numeric values:", non_numeric_columns)
```

```python
# Assuming 'url' is your DataFrame and 'columns_to_remove' is a list of column names you want to remove
columns_to_remove = ['tld_params','extension']  # Example list of columns to remove

# Removing the specified columns
url.drop(columns=columns_to_remove, inplace=True)
```

```python
url.dtypes
```

```python
import pandas as pd

# Assuming df is your DataFrame

# Identify boolean columns
bool_columns = url.select_dtypes(include=bool).columns

# Exclude boolean columns and the column named 'URL'
columns_to_change = [col for col in url.columns if col not in bool_columns and col != 'URL']

# Change data types of selected columns
url[columns_to_change] = url[columns_to_change].astype('float32')  # Change 'new_data_type' to the desired data type

```

```python
url.dtypes
```

```python
X = url.iloc[:, :-2]
Y = url.iloc[:, -2]
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split as tts
xtrain,xtest,ytrain,ytest=tts(X,Y,test_size=0.25,random_state=6)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Assuming you have already split your data into training and testing sets: xtrain, xtest, ytrain, ytest

# Create a logistic regression model
logistic_regression = LogisticRegression()

# Train the model on the training data
logistic_regression.fit(xtrain, ytrain)

# Make predictions on the test data
predictions = logistic_regression.predict(xtest)

# Evaluate the model's accuracy
accuracy = accuracy_score(ytest, predictions)
print("Accuracy:", accuracy)
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

# Instantiate the LabelEncoder to encode the categorical target variable
label_encoder = LabelEncoder()
ytrain_encoded = label_encoder.fit_transform(ytrain)
ytest_encoded = label_encoder.transform(ytest)

# Instantiate the XGBoost classifier
model = xgb.XGBClassifier()

# Train the model
model.fit(xtrain, ytrain_encoded)

# Make predictions on the test set
predictions = model.predict(xtest)

# Evaluate the model
accuracy = accuracy_score(ytest_encoded, predictions)
print("Accuracy:", accuracy)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score

# Instantiate the AdaBoost classifier
model = AdaBoostClassifier()

# Train the model
model.fit(xtrain, ytrain)

# Make predictions on the test set
predictions = model.predict(xtest)

# Evaluate the model
accuracy = accuracy_score(ytest, predictions)
print("Accuracy:", accuracy)
```
